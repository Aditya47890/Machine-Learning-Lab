```python
class MeraLR:

  def __init__(self):
    self.coef_ = None
    self.b = None

  def fit(self, X_train, y_train):
    # X_train.shape
    num = 0
    den = 0
    for i in range(X_train.shape[0]):
      num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
      den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))

    self.coef_ = num/den
    self.b = y_train.mean()- (self.coef_ * X_train.mean())
    print(self.coef_, self.b)
    return self.coef_, self.b



  def predict(self,X_test):
    # pass
    return self.coef_ * X_test + self.b
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace path with your actual CSV file if needed)
df = pd.read_csv("advertising.csv")

# Extract TV and Sales columns
X = df['TV'].values
Y = df['Sales'].values

# Normalize features (optional but useful)
X = (X - np.mean(X)) / np.std(X)
Y = (Y - np.mean(Y)) / np.std(Y)
```

```python
#Batch GD
```

```python
def batch_gradient_descent(X, Y, alpha=0.01, epochs=1000):
    n = len(X)
    w, b = 0.0, 0.0
    loss_history = []

    for epoch in range(epochs):
        y_pred = w * X + b
        error = Y - y_pred

        # Gradients
        dw = (-2/n) * np.sum(X * error)
        db = (-2/n) * np.sum(error)

        # Update
        w -= alpha * dw
        b -= alpha * db

        # MSE loss
        loss = np.mean(error**2)
        loss_history.append(loss)

        if epoch % 100 == 0:
            print(f'Epoch {epoch}: Loss={loss:.4f}, w={w:.4f}, b={b:.4f}')

    return w, b, loss_history
```

```python
#SGD
```

```python
def stochastic_gradient_descent(X, Y, alpha=0.01, epochs=1000):
    n = len(X)
    w, b = 0.0, 0.0
    loss_history = []

    for epoch in range(epochs):
        total_loss = 0
        for i in range(n):
            xi = X[i]
            yi = Y[i]
            y_pred = w * xi + b
            error = yi - y_pred

            # Gradients
            dw = -2 * xi * error
            db = -2 * error

            # Update
            w -= alpha * dw
            b -= alpha * db

            total_loss += error**2

        loss = total_loss / n
        loss_history.append(loss)

        if epoch % 100 == 0:
            print(f'Epoch {epoch}: Loss={loss:.4f}, w={w:.4f}, b={b:.4f}')

    return w, b, loss_history
```

```python
def plot_loss(loss_history, title):
    plt.figure(figsize=(8, 4))
    plt.plot(loss_history, label='Loss')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Batch GD
w_bgd, b_bgd, loss_bgd = batch_gradient_descent(X, Y, alpha=0.01, epochs=1000)
plot_loss(loss_bgd, 'Batch Gradient Descent Loss')

# Online GD
w_sgd, b_sgd, loss_sgd = stochastic_gradient_descent(X, Y, alpha=0.01, epochs=1000)
plot_loss(loss_sgd, 'Stochastic Gradient Descent Loss')
```
